```python
import sys
sys.path.insert(0, '.')
from host_discovery import craft_discovery_pkts, discover_hosts
```

```python
# Host activo - TCP ACK
discover_hosts('tcp_ack', '172.28.0.10', port=80)
```

```python
# Host activo - ICMP Timestamp
discover_hosts('icmp_timestamp', '172.28.0.11')
```

```python
# Host inactivo - sin respuesta esperada
discover_hosts('udp', '172.28.0.99', port=80)
```

```python
# Usando los 3 protocolos a la vez sobre host activo
discover_hosts(
    protocols=['tcp_ack', 'udp', 'icmp_timestamp'],
    ip_range='172.28.0.11',
    pkt_count={'tcp_ack': 2, 'udp': 1, 'icmp_timestamp': 1},
    port=80
)
```
